Use TF-IDF to find some related books

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings

warnings.filterwarnings('ignore')

sns.set()
```

```python
books = pd.read_csv('books.csv')
```

```python
books.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df = 0, stop_words = 'english')
tfidf = tf.fit_transform(books['authors'])
tfidf
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine = linear_kernel(tfidf, tfidf)
cosine
```

```python
title = books['title']
title_df = pd.Series(books.index, index=books['title'])
title_df['1984']
```

```python
cosine[title_df['1984']]
```

```python
c_scores = list(enumerate(cosine[title_df['1984']]))
c_scores = sorted(c_scores, key = lambda x : x[1], reverse= True)
c_scores[:3]
```

```python
print(f'Index 6번의 책 이름 :', books['title'][12])
print(f'Index 18번의 책 이름 :', books['title'][13])
print(f'Index 154번의 책 이름 :', books['title'][2583])
```

```python
sim_scores = c_scores[1:11]
book_indices = [i[0] for i in sim_scores]
title.iloc[book_indices]
```
